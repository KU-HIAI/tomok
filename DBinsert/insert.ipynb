```python
from urllib.parse import urlparse
from langchain.vectorstores.pgvector import PGVector
import psycopg2

def connect_simple():
    url = PGVector.connection_string_from_db_params(
        driver='psycopg2',
        host=ORM_HOST,
        port=ORM_PORT,
        database=ORM_DATABASE,
        user=ORM_USERNAME,
        password=ORM_PASSWORD
    )
    result = urlparse(url)
    conn = psycopg2.connect(
        database=result.path[1:],
        user=result.username,
        password=result.password,
        host=result.hostname,
        port=result.port
    )
    return conn
```

```python
import os

def find_py_files_and_content(root_dir):
    files_and_contents = []
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith('.py'):
                # 파일 경로에서 root 디렉토리를 제거
                relative_path = os.path.relpath(os.path.join(root, file), root_dir)
                # 파일 이름에서 .py 확장자 제거
                base_name = os.path.splitext(file)[0]
                # 경로를 디렉토리와 파일 이름으로 분할
                directories = relative_path.split(os.sep)[:-1]  # 파일 이름 제외
                directories.append(base_name)

                # 파일 내용 읽기
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()

                # 경로와 내용을 튜플로 묶어 리스트에 추가
                files_and_contents.append((directories, content))

    return files_and_contents
# 사용 예
root_directory = '../ruleunits'  # 여기에 원하는 디렉토리 경로를 넣으세요.
py_paths_contents  = find_py_files_and_content(root_directory)[1:]


def build_classname(dir_list):
    return dir_list[0].upper()+dir_list[1]+'_'+dir_list[2]

# 출력 예
for path, contents in py_paths_contents[:3]:
    if len(path) > 2:
        print(path)
        print(build_classname(path))
        print(contents)
```

```python
def get_values(ru):
    fc = ru.flowchart
    metadata = {}
    metadata['priority'] = ru.priority
    metadata['version'] = ru.version
    metadata['author'] = ru.author
    metadata['ref_code'] = ru.ref_code
    metadata['ref_date'] = ru.ref_date
    metadata['doc_date'] = ru.doc_date
    metadata['title'] = ru.title
    metadata['description'] = ru.description
    content = ru.content
    
    return fc, metadata, content
```

```python
controller = RuleUnitController(path='../ruleunits')

values = []
contents = []

for path, content in py_paths_contents:
    if len(path) > 2:
        try:
            values.append(get_values(
                controller[path[0]][path[1]][path[2]][build_classname(path)]
            ))
            contents.append(content)
        except:
            continue
```

```python
def get_function_info(content: str):
    lines = content.split('\n')
    
    f_string = ""
    flag = False
    
    for line in lines[:-3]:
        if flag:
            f_string += "\n" + line
    
        if "@rule_method" in line:
            flag = True
    
    f_string = f_string.strip()
    # print(f_string)

    lines = f_string.split('\n')
    function_def = lines[0].split(' ')[1].split('(')
    function_name = function_def[0]
    function_vars = function_def[1][:-1]
    
    return f_string, function_name, function_vars

get_function_info(contents[0])
```

```python
import json
import re


for i in range(len(values)):
    try:
        conn = connect_simple()
        cur = conn.cursor()
        
        ref_code = re.sub(r" \(\d+\)", "", values[i][1]['ref_code'])
        
        cur.execute("SELECT std_id FROM standard_item WHERE code = %s", (ref_code,))
        std_id_result = cur.fetchone()
        
        if std_id_result:
            std_id = std_id_result[0]
            
            insert_query = """INSERT INTO std_rule_unit (rule_unit_id, std_id, py_code, flowchart, metadata, content) VALUES 
            (uuid_generate_v4(), %s, %s, %s, %s, %s)
            RETURNING rule_unit_id;
            """
            cur.execute(insert_query, (std_id, contents[i], values[i][0], json.dumps(values[i][1], ensure_ascii=False), values[i][2]))
            
            rule_unit_id = cur.fetchone()[0]
            
            insert_query = """INSERT INTO std_function (func_id, rule_unit_id, name, source, api_spec, std_id) VALUES
            (uuid_generate_v4(), %s, %s, %s, %s, %s)
            RETURNING func_id
            """
            source, name, _ = get_function_info(contents[i])
            
            cur.execute(insert_query, (rule_unit_id, name, source, None, std_id))
            
            func_id = cur.fetchone()[0]
            
            conn.commit()
        else:
            insert_query = """INSERT INTO std_rule_unit (rule_unit_id, std_id, py_code, flowchart, metadata, content) VALUES 
            (uuid_generate_v4(), uuid_generate_v4(), %s, %s, %s, %s)
            RETURNING rule_unit_id;
            """
            cur.execute(insert_query, (contents[i], values[i][0], json.dumps(values[i][1], ensure_ascii=False), values[i][2]))
            
            rule_unit_id = cur.fetchone()[0]
            
            insert_query = """INSERT INTO std_function (func_id, rule_unit_id, name, source, api_spec, std_id) VALUES
            (uuid_generate_v4(), %s, %s, %s, %s, %s)
            RETURNING func_id
            """
            source, name, _ = get_function_info(contents[i])
            
            cur.execute(insert_query, (rule_unit_id, name, source, None, std_id))
            
            func_id = cur.fetchone()[0]
            
            conn.commit()
    except:
        print(i)
    finally:
        cur.close()
        conn.close()
```
